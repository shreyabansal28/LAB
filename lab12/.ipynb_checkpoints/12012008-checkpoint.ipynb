```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import seaborn as sn
from sklearn.model_selection import train_test_split
from sklearn.utils import resample
import matplotlib.pyplot as plt
df = pd.read_csv('customerspends.csv')
df.info()
```

```python
# 1. Create a scatter plot to visualise dataset 
plt.figure(figsize=(8, 6))
plt.scatter(df['Apparel'], df['Beauty and Healthcare'])
plt.xlabel('Apparel Spending (in INR)')
plt.ylabel('Beauty and Healthcare Spending (in INR)')
plt.title('Customer Spending on Apparel and Beauty/Healthcare Products')
plt.grid(True)
plt.show()

# As we can see in the scatter plot there are 3 possible clusters in the dataset
```

```python
# 2 Normalize the features using StandardScaler and plot them in the scatter plot again.
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = df[['Apparel', 'Beauty and Healthcare']]
scaled_df = scaler.fit_transform(X);
plt.figure(figsize=(8, 6))
plt.scatter(scaled_df[:,0], scaled_df[:,1])
plt.xlabel('Apparel Spending (in INR)')
plt.ylabel('Beauty and Healthcare Spending (in INR)')
plt.title('Customer Spending on Apparel and Beauty/Healthcare Products')
plt.grid(True)
plt.show()
```

```python
# 3 Use dendrogram and elbow method to verify if the number of clusters suggested is same as clusters visible in the scatter plot in Question 1
cmap = sn.cubehelix_palette(as_cmap=True, rot=-.3, light=1)
sn.clustermap( scaled_df, cmap=cmap, linewidths=.2,
figsize = (8,8));

# By analysing below plot we can say that number of clusters are same as of question 1 that is 3
```

```python
# 4 Create the number of clusters as suggested by the elbow method using K-means algorithm.
from sklearn.cluster import KMeans
cluster_range = range(1, 10)
cluster_errors = []
for num_clusters in cluster_range:
    clusters = KMeans(num_clusters, n_init="auto")
    clusters.fit(scaled_df)
    cluster_errors.append(clusters.inertia_)
plt.figure(figsize=(6,4))
plt.plot(cluster_range, cluster_errors, marker = "o");
```

```python

# By analysing the below elbow curve we can see that the elbow is found at n = 3 so number of clusters are 3
k = 3
clusters = KMeans(k, random_state = 42, n_init="auto")
clusters.fit(scaled_df)
df["clusterid"] = clusters.labels_

# 5. Print the records of customers in each cluster and the cluster center of each cluster separately. Explain the clusters intuitively
# 1 printing the elements in cluster 0
df[df.clusterid == 0]
```

```python
# 2 printing the elements in cluster 1
df[df.clusterid == 1]
```

```python
# 3. printing the elements in cluster 2
df[df.clusterid == 2]
```

```python
# 4. Printing the cluster centers
cluster_centers = clusters.cluster_centers_
print("Cluster centers are : \n", cluster_centers)
```

```python
# Q.6  Use Online Retail.xlsx Select only the transactions that have occurred in the year 06/01/11 and 11/30/11 and create a new subset of data. For answering questions from 7 to 10, use this subset of data.

data = pd.read_excel("Online Retail.xlsx")
data.info()
```

```python
# splitting the dataset according to the given date 
data["InvoiceDate"] = pd.to_datetime(data["InvoiceDate"])

# Define the start and end date for the subset
start_date = pd.to_datetime("2011-06-01")
end_date = pd.to_datetime("2011-11-30")

# Create a subset of data for the specified date range
subset_data = data[(data["InvoiceDate"] >= start_date) & (data["InvoiceDate"] <= end_date)]
# print(subset_data.size)
```

```python
# 7. Calculate the RFM values for each customer (by customer id). RFM represents
from datetime import datetime

# Define the reference date for Recency
reference_date = datetime(2011, 11, 30)

# Calculate Recency
recency = data.groupby("CustomerID")["InvoiceDate"].max().reset_index()
recency["Recency"] = (reference_date - recency["InvoiceDate"]).dt.days // 30

# Calculate Frequency
frequency = data.groupby("CustomerID")["InvoiceNo"].nunique().reset_index()
frequency.columns = ["CustomerID", "Frequency"]

# Calculate Monetary Value
data["TotalPrice"] = data["Quantity"] * data["UnitPrice"]
monetary = data.groupby("CustomerID")["TotalPrice"].sum().reset_index()
monetary.columns = ["CustomerID", "Monetary"]

# Merge the Recency, Frequency, and Monetary DataFrames
rfm_data = recency.merge(frequency, on="CustomerID").merge(monetary, on="CustomerID")

# Display the RFM data for each customer
print(rfm_data.head(5))
```

```python
# 8. Use dendrogram and elbow method to identify how many customer segments exist, using the RFM values for each customer created in Question 7
rfm_data = rfm_data.drop(["InvoiceDate"], axis=1)
scaler = StandardScaler()
X = rfm_data[['CustomerID', 'Recency', 'Frequency', 'Monetary']]
scaled_rfm_df = scaler.fit_transform(X)
cmap = sn.cubehelix_palette(as_cmap=True, rot=-.3, light=1)
sn.clustermap( scaled_rfm_df, cmap=cmap, linewidths=.2,
figsize = (8,8));
```

```python
# 8th elbow method 
from sklearn.cluster import KMeans
cluster_range = range(1, 10)
cluster_errors = []
for num_clusters in cluster_range:
    clusters = KMeans(num_clusters, n_init="auto")
    clusters.fit(scaled_rfm_df)
    cluster_errors.append(clusters.inertia_)
plt.figure(figsize=(6,4))
plt.plot(cluster_range, cluster_errors, marker = "o")
# By analysing the above graph we can see that number of clusters can be either 4 or 5
```

```python
# 9. Create the customer segments with K-means algorithm by using number of clusters is suggested by elbow method.
k = 4
clusters = KMeans(k, random_state = 42, n_init="auto")
clusters.fit(scaled_rfm_df)
rfm_data["clusterid"] = clusters.labels_

# a. Printing the elements in cluster 0
rfm_data[rfm_data.clusterid == 0]
```
