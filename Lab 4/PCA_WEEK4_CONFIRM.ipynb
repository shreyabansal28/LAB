```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from numpy.linalg import eig



df = pd.read_csv('stud2.csv')


features = ['Marks', 'DSA', 'soft_skill', 'engg skills']

output = ['Bond-months']


x = df.loc[:, features].values
y = df.loc[:, output].values

x = StandardScaler().fit_transform(x)


x_df = pd.DataFrame(x, columns = features)


## save x_df to csv
x_df.to_csv('12012002_08_inp.csv', index=False)

## covariance matrix
covar_matrix = x_df.cov()

## eigen values and eigen matrix
w, v = eig(covar_matrix)

## w [ 2.88772767e+00  7.60556958e-01  4.80747634e-01 -3.39580439e-16]

print('w', w)

# print('x: ', x)


from sklearn.decomposition import PCA

pca = PCA(n_components=2)

principalComponents = pca.fit_transform(x)


## calculating only first two PCs based on two highest eigen values
principalDf = pd.DataFrame(data = principalComponents, columns = ['pc1', 'pc2'])


## save principalDf to csv
principalDf.to_csv('12012002_08_out.csv', index=False)

print('principalDf: ', principalDf)
```
