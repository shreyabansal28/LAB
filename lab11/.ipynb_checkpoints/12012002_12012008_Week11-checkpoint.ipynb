```python
import pandas as pd


## check for data imbalance
## conditon -- True/False < 33%
## upscaling logic on pg 212-213 of book

def calculate_percentages(hr_data_df, col_name):
    value_counts = hr_data_df[col_name].value_counts()

    ### calculating percentages
    joined_cnt = value_counts['Joined']
    not_joined_cnt = value_counts['Not Joined']

    frac_joined = (joined_cnt)/(joined_cnt + not_joined_cnt)
    frac_not_joined = (not_joined_cnt)/(joined_cnt + not_joined_cnt)
    
    return {"total_samples": (joined_cnt + not_joined_cnt), "fraction_joined" : frac_joined, "fraction_not_joined": frac_not_joined }


hr_data_df = pd.read_csv('hr_data.csv')
percent_calcs = calculate_percentages(hr_data_df, 'Status')

print("total samples: ", percent_calcs['total_samples'])
print("fraction_joined ", percent_calcs['fraction_joined'])
print("fraction not joined: ", percent_calcs['fraction_not_joined'])
```

```python
## since frac not joined is 0.18 we need to upscale the data
from sklearn.utils import resample, shuffle

joined_records = hr_data_df[hr_data_df.Status == 'Joined']
not_joined_records = hr_data_df[hr_data_df.Status == 'Not Joined']

### upsampling the 'Not Joined' cases 
hr_data_df_upsampled = resample(not_joined_records, replace=True,
                               n_samples=6000)

upsampled_df = pd.concat([joined_records, hr_data_df_upsampled]) ## this is the final df to be used throughout

## calculating % of new df
percent_calcs = calculate_percentages(upsampled_df, 'Status')

print("total samples: ", percent_calcs['total_samples'])
print("fraction_joined ", percent_calcs['fraction_joined'])
print("fraction not joined: ", percent_calcs['fraction_not_joined'])
```

```python
## creating X_features

X_features = list(upsampled_df.columns)
features_to_remove = ['Status', 'SLNO', 'Candidate Ref']
for feature in features_to_remove:
    X_features.remove(feature)

    
## getting dummies

encoded_df = pd.get_dummies(upsampled_df[X_features], drop_first=True)

X = encoded_df
Y = upsampled_df.Status.map(lambda x: int( x == 'Joined'))

print("X: ", X)
print("Y: ", Y)
```

```python
## creating train/test split
from sklearn.model_selection import train_test_split

train_X, test_X, train_Y, test_Y = train_test_split(X, Y, test_size=0.3, random_state=42)

train_X
```

```python
from sklearn import metrics
import matplotlib.pyplot as plt

def draw_roc_curve( model, test_X, test_y ):
    ## Creating and initializing a results DataFrame with actual
    ## labels
    test_results_df = pd.DataFrame( { 'actual': test_y } )
    test_results_df = test_results_df.reset_index()
    # predict the probabilities on the test set
    predict_proba_df = pd.DataFrame( model.predict_proba( test_X.values ) )
    ## selecting the probabilities that the test example belongs
    ## to class 1
    test_results_df['chd_1'] = predict_proba_df.iloc[:,1:2]
    ## Invoke roc_curve() to return fpr, tpr and threshold values.
    ## Threshold values contain values from 0.0 to 1.0
    fpr, tpr, thresholds = metrics.roc_curve( test_results_df.actual,
    test_results_df.chd_1,
    drop_intermediate = False )
    ## Getting roc auc score by invoking metrics.roc_auc_score method
    auc_score = metrics.roc_auc_score( test_results_df.actual,
    test_results_df.chd_1 )
    ## Setting the size of the plot
    plt.figure(figsize=(8, 6))
    ## Plotting the actual fpr and tpr values
    plt.plot(fpr, tpr, label = 'ROC curve (area = %0.2f)' % auc_score)
    ## Plotting th diagnoal line from (0,1)
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    ## Setting labels and titles
    plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
```

```python
import seaborn as sn

def draw_cm(actual, predicted):
    confusion_matrix = metrics.confusion_matrix(actual, predicted)
    sn.heatmap(confusion_matrix, annot=True, fmt='.2f', xticklabels=['Joined', 'Not Joined'], yticklabels=['Joined', 'Not Joined'])
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()
```

```python
## Building SVM model using linear kernel

from sklearn.svm import SVC

svm_clf = SVC(kernel='linear', C=1.0, probability=True) 
svm_clf.fit(train_X, train_Y)
```

```python
## drawing ROC curve for the model with "linear" kernel

_, _, _, _ = draw_roc_curve(svm_clf, test_X, test_Y)
```

```python
## analysis of SVM model with linear kernel using confusion matrix

pred_Y_svm_clf = svm_clf.predict(test_X.values)
draw_cm(test_Y, pred_Y_svm_clf)
```

```python
# analysis using model summary
print(metrics.classification_report(test_Y, pred_Y_svm_clf))
```
