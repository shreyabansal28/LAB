```python
import matplotlib.pyplot as plt
import seaborn as sn
import pandas as pd
bollywood_data_set = pd.read_csv('bollywood.csv')
bollywood_data_set.info()
```

```python
roi = []
for i in range(0, len(bollywood_data_set)):
    data_set = bollywood_data_set.iloc[i]
    rate = (data_set['BoxOfficeCollection'] - data_set['Budget'])/data_set['Budget']
    roi.append(rate)
bollywood_data_set['Rate of return'] = roi
bollywood_data_set.head(5)
```

```python
sn.distplot (bollywood_data_set[bollywood_data_set['Genre'] == 'Comedy']
['Rate of return'],
color = 'y',
label = 'Comedy movies')
sn.distplot (bollywood_data_set[bollywood_data_set['Genre'] == ' Drama ']
['Rate of return'],
color = 'r',
label = 'Drama movies')
plt.legend();
```
