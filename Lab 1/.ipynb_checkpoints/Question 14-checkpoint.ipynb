```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv('SAheart.data')
data.info()
```

```python
chd_famhist_count = data.groupby(["chd", "famhist"]).size().unstack()
# print(chd_famhist_count)

# Plotting using seaborn barplot
sns.barplot(data=chd_famhist_count)
plt.title("Number of Persons with CHD based on Family History")
plt.xlabel("Family History of Heart Disease")
plt.ylabel("Number of Persons")
plt.legend(title="CHD", labels=["No CHD", "CHD"])
plt.show()
```
