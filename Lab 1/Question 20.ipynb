```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
heart_db = pd.read_csv('SAheart.data')
heart_db.info()
```

```python
def find_age_group(age):
    if age <= 15 :
        return "young"
    elif age <= 35:
        return "adults"
    elif age <= 55: 
        return "mid"
    else:
        return "old"

heart_db['age_group'] = heart_db['age'].apply(find_age_group)
sns.boxplot(data = heart_db, x = "age_group", y ="ldl")
```
