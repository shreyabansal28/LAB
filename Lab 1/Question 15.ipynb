```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
heart_db = pd.read_csv('SAheart.data')
heart_db.info()
```

```python
# Creating a scatter plot

sns.scatterplot(data=heart_db, x="age", y="sbp")
plt.title("Scatter Plot of Age vs Systolic Blood Pressure")
plt.xlabel("Age")
plt.ylabel("Systolic Blood Pressure")
plt.show()

# Calculating correlation coefficient
correlation = heart_db["age"].corr(heart_db["sbp"])
print(f"Correlation coefficient between age and sbp: {correlation:.2f}")
```
